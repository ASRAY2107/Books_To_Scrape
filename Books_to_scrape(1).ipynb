**In This Project Have Scraped The important Data From The Website And Have Performed a Analysis To Know The Relation Between Prices And Ratings**

**WEBSITE LINK** - http://books.toscrape.com/catalogue/page-1.html

```python
pages = []
prices = []
Rating = []
urls = []
titles = []

no_of_pages = 1;

for i in range(1,no_of_pages+1):
  url = ('http://books.toscrape.com/catalogue/page-{}.html'.format(i))
  pages.append(url);
```

```python
print("number of pages",len(pages))

print(pages)
```

```python
import requests
import bs4
for item in pages:
  page = requests.get(item)
  soup = bs4.BeautifulSoup(page.text,'html.parser')
print(soup)
```

```python


print(soup.prettify())
```

```python

print("All the Titles in the page 1 : ")

for t in soup.findAll('h3'):
  print(t)
```

```python
for t in soup.findAll('h3'):
  title = t.getText()
  titles.append(title)
```

```python
titles
```

```python
for price in soup.findAll('p' , class_ = 'price_color'):
  p = price.getText()
  prices.append(p)
```

```python
prices
```

```python
for s in soup.findAll('p',class_ = 'star-rating'):
  for k,v in s.attrs.items():
    stars = v[1]
    Rating.append(stars)
```

```python
Rating
```

```python
divs=soup.find_all('div',class_='image_container')
for thumbs in divs:
    tagss=thumbs.find('img',class_='thumbnail')
    links='http://books.toscrape.com/' + str(tagss['src'])
    newlinks=links.replace('..','')
    urls.append(newlinks)
```

```python
urls
```

```python
web_datas = {'title':titles,'Ratings':Rating,'urls':urls,'prices':prices}
```

```python
print(len(titles))
print(len(Rating))
print(len(urls))
print(len(prices))
```
